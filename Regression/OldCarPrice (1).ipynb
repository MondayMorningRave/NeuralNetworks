```python
df["car_prices_in_rupee"] = df["car_prices_in_rupee"].str.extract(pattern)
df["car_prices_in_rupee"] = pd.to_numeric(df["car_prices_in_rupee"], errors="coerce")
df["car_prices_in_rupee"].isna().sum()
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
```

```python
df['kms_driven'] = scaler.fit_transform(df[['kms_driven']])
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal = OrdinalEncoder()
```

```python
oe_features = ['car_name', 'fuel_type', 'transmission', 'ownership', 'manufacture','engine','seats']
for feature in oe_features:
    df[feature] = ordinal.fit_transform(df[[feature]])
```

```python
X = df.drop(["unnamed:_0", "car_prices_in_rupee"], axis=1)
y = df["car_prices_in_rupee"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.20, shuffle=True, random_state=1)
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import LearningRateScheduler
from tensorflow.keras.losses import MeanSquaredError
from sklearn.metrics import mean_squared_error
```

```python
def create_model(input_shape, dropout_rate=0.2, weight_decay=0.0001):
    model = Sequential()
    model.add(layers.InputLayer(input_shape=input_shape))
    model.add(layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(weight_decay)))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(weight_decay)))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(weight_decay)))
    model.add(layers.Dropout(dropout_rate))
    model.add(layers.Dense(1, activation='linear'))
    return model
```

```python
def scheduler(epoch, lr):
    if epoch < 30:
        return lr
    else:
        return lr * 0.1
```

```python
input_shape = X_train.shape[1]
model = create_model(input_shape)
model.compile(optimizer=Adam(learning_rate=0.001), loss=MeanSquaredError(), metrics=['mse'])

callbacks = [LearningRateScheduler(scheduler, verbose=1)]
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_val, y_val), callbacks=callbacks)
```
